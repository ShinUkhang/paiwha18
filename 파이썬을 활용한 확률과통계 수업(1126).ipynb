# 3. 정규분포곡선

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

f=open('서울시기온정보.csv')
data=csv.reader(f)
next(data)

result=[]
#11월
for row in data:
    if row[-3] != '' :
        result.append(float(row[-3]))
#평균과 분산
mu=np.mean(result)
sigma=np.std(result)
#정규분포곡선
x=np.linspace(-15,40)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--', color='g')
plt.show()
```

# 7.자료의 분석 및 추론

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rc('font',family='Malgun Gothic')
plt.figure(dpi=100)

f=open('서울시기온정보.csv')
data=csv.reader(f)
next(data) #두번째 줄부터 데이터로 인식

result=[]
for row in data:
    if row[-3] != '' :
        result.append(float(row[-3]))
#평균과 분산
mu=np.mean(result)
sigma=np.std(result)
#히스토그램
plt.hist(result, bins=40, density=True, edgecolor='k', linewidth=1.2, alpha=0.4)

#정규분포
x=np.linspace(-10,40)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--', color='g')
plt.title('110년간 평균기온과 정규분포')
plt.savefig('110년 기온분포')
plt.show()
print(mu,sigma)
```

# 8.가을(11월) 기온분석

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rc('font',family='Malgun Gothic')

f=open('서울시기온정보.csv')
data=csv.reader(f)
next(data) 

result=[]
#11월 달의 기온을 110년간 추출
for row in data:
    if row[-3] != '' and row[0].split('-')[1] == '11' :
        result.append(float(row[-3]))
#평균과 분산
mu=np.mean(result)
sigma=np.std(result)
#히스토그램
plt.hist(result, bins=40, density=True, edgecolor='k', linewidth=1.2, alpha=0.4)

#정규분포
x=np.linspace(-10,20)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--', color='g')
plt.title('11월 기온과 정규분포')
plt.savefig('11월 기온분포')
plt.show()
print(mu,sigma)
```
