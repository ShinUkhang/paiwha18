```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.stats import norm
plt.rcParams['axes.unicode_minus']=False
plt.rc('font', family='HYGothic-Extra')

#1년간 주가분석(KoSPI 종가)
r1=[]
f=open('kospi2018.csv', encoding='utf8')
data=csv.reader(f)

for row in data:
     if '종가' not in row:
            r1.append(float(row[1].replace(',','')))

#히스토그램으로 그리기
plt.hist(r1, bins=30,density=False, edgecolor='k', linewidth=1.2)
plt.title('KOSPI주가분석')
plt.show()


#이항분포로 그리기
sm=norm.rvs(size=10000)
cnt,b,ig=plt.hist(sm, 50,density=False, edgecolor='k', linewidth=1.2 )
plt.title('이항분포')
plt.show()

#정규분포로 그리기
plt.plot(b,norm.pdf(b), color='r')
plt.title('정규분포')
a,b=norm.interval(0.95)
plt.axvline(x=a, linestyle='--', label=round(a,2))
plt.axvline(x=b, linestyle='--', label=round(b,2))
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
from scipy.stats import norm
plt.rcParams['axes.unicode_minus']=False
plt.rc('font', family='HYGothic-Extra')

#1년간 주가분석(KoSPI 종가)
r1=[]
f=open('kospi2018.csv', encoding='utf8')
data=csv.reader(f)

for row in data:
     if '종가' not in row:
            r1.append(float(row[1].replace(',','')))

#히스토그램으로 그리기
plt.subplot(3,1,1)
plt.hist(r1, bins=30,density=False, edgecolor='k', linewidth=1.2)
plt.title('KOSPI주가분석')
plt.show()


#이항분포로 그리기
plt.subplot(3,1,2)
sm=norm.rvs(size=10000)
cnt,b,ig=plt.hist(sm, 50,density=False,color='g', edgecolor='k', linewidth=1.2 )
plt.title('이항분포')
plt.show()

#정규분포로 그리기
plt.subplot(3,1,3)
plt.plot(b,norm.pdf(b), color='r')
plt.title('정규분포')
a,b=norm.interval(0.95)
plt.axvline(x=a, linestyle='--', label=round(a,2))
plt.axvline(x=b, linestyle='--', label=round(b,2))
plt.legend()
plt.savefig('과제예시')
plt.show()
```
