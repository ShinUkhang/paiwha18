```python
import csv
f=open("지하철통계_1807.csv")
data=csv.reader(f)
next(data)


line=input("지하철 노선을 입력하세요(예: 1호선)")

r1=[]
r2=[]
n=0
for row in data:
    if line in row[1]:
        r1.append(row[3])
        r2.append(int(row[4].replace(",","")))
        if int(row[4].replace(",",""))>=n:
            n=int(row[4].replace(",",""))
            m_stn=row[3]

import matplotlib.pyplot as plt
plt.plot(r1,r2,"bs-")
plt.title(line)
plt.figuresize=(16,4)
plt.style.use("ggplot")
plt.rc("font", family="Malgun Gothic")
plt.rc("xtick",labelsize=4)
plt.figure(figsize=(4,3),dpi=1000)
print(line, "에서 가장 승차객이 많은 역은",m_stn,"역입니다.")
plt.savefig("line.png")
plt.show()


        
```
