```python
import csv
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

#데이터 불러오기
f=open('전국cctv표준데이터(숫자).csv')
data = csv.reader(f)

lat=[]
lon=[]
camera=[]

for row in data:
    if int(row[0])>2 and (float(row[1])<=38) and (float(row[2])<=130):
        lat.append(float(row[1]))
        lon.append(float(row[2]))
        camera.append(int(row[0]))

plt.figure(figsize=(10,10))
plt.scatter(lon,lat)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

#데이터 불러오기
f=open('전국cctv표준데이터(숫자).csv')
data = csv.reader(f)

lat=[]
lon=[]
camera=[]

for row in data:
    if int(row[0])>2 and (float(row[1])<=38) and (float(row[2])<=130):
        lat.append(float(row[1]))
        lon.append(float(row[2]))
        camera.append(int(row[0]))
size=np.array(camera)*3
colormap=np.array(camera)*2
color_num=7
plt.figure(figsize=(10,10))
plt.scatter(lon,lat,s=size,c=colormap, cmap=plt.cm.get_cmap('rainbow', color_num), alpha=0.2)
plt.colorbar(ticks=range(color_num), format='color: %d', label='color')
plt.show()
```
