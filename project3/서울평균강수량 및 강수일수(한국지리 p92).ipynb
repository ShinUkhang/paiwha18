# 8. 비율로 계산해서 한 그래프에 그려본다. 

```python
import csv
import matplotlib.pyplot as plt

f=open('2017서울시평균강수량.csv')
data=csv.reader(f)

r1=[]
r2=[]
r3=[]
for row in data:
    for i in range (3,(len(row))):
        if '기간' in row:
            r1.append(row[i])
        if '강수량' in row:
            r2.append(float(row[i])/float(row[2].replace(',', '')))
        if '강수일수' in row:
            r3.append(float(row[i])/float(row[2].replace(',', '')))
plt.rc('font', family='Malgun Gothic')
plt.bar(r1,r2)
plt.plot(r1,r3,color='r', marker='x', linestyle='--')
plt.show()
```

# 10. 그래프를 꾸며본다.

```python
import csv
import matplotlib.pyplot as plt

f=open('2017서울시평균강수량.csv')
data=csv.reader(f)

r1=[]
r2=[]
r3=[]
for row in data:
    for i in range (3,(len(row))):
        if '기간' in row:
            r1.append(row[i])
        if '강수량' in row:
            r2.append(float(row[i]))
        if '강수일수' in row:
            r3.append(float(row[i]))
plt.rc('font', family='Malgun Gothic')
fig=plt.figure(figsize=(5,5))
ax1=fig.add_subplot(111)
ax1.plot(r1,r2, 'o-r',label='강수량')
ax1.legend(loc=1)
ax2=ax1.twinx()
ax2.bar(r1,r3,color='b',width=0.5,alpha=0.4, label='강수일수')
ax2.legend(loc=2)
plt.title('2017 서울시강수량 및 강수일수비교')
plt.show()
```

# 13 선형회귀 및 상관계수 분석 

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
f=open('2017서울시평균강수량.csv')
data=csv.reader(f)

r1=[]
r2=[]
r3=[]
for row in data:
    for i in range (3,(len(row))):
        if '기간' in row:
            r1.append(row[i])
        if '강수량' in row:
            r2.append(float(row[i]))
        if '강수일수' in row:
            r3.append(float(row[i]))
plt.rc('font', family='Malgun Gothic')

r3=pd.Series(r3)
r2=pd.Series(r2)
ax=plt.subplots()
ax=sns.regplot(r2,r3)
np.corrcoef(r2,r3)
```

# 만화처럼 그려보기

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd

f=open('2017서울시평균강수량.csv')
data=csv.reader(f)

r1=[]
r2=[]
r3=[]
for row in data:
    for i in range (3,(len(row))):
        if '기간' in row:
            r1.append(row[i])
        if '강수량' in row:
            r2.append(float(row[i]))
        if '강수일수' in row:
            r3.append(float(row[i]))
plt.rc('font', family='Malgun Gothic')

plt.xkcd()
r2=pd.Series(r2)

plt.annotate(s='Peak', xy=(r2.argmax(),r2.max()),xytext=(r2.argmax()+0.5,r2.max()+0.1), arrowprops={'facecolor':'black', 'shrink':0.1})

plt.plot(r1,r2)
plt.show()
```
