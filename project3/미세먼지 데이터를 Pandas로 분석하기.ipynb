# 2. 그래프로 자료표현하기(단위가 같은 경우)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

#데이터 불러오기
f=open('대기정보측정자료_180825.csv')
data=pd.read_csv(f,index_col=0, parse_dates=['날짜'])

fig=plt.figure(figsize=(10,4))
ax=fig.add_subplot(111)
ax.plot(data.index, data['오존'], label='오존', color='red')
ax.plot(data.index, data['일산화탄소'], label='일산화탄소')
ax.plot(data.index, data['이산화질소'], label='이산화질소')
ax.plot(data.index, data['아황산가스'], label='아황산가스')


ax.set_title('대기정보측정자료')
ax.set_ylabel('ppm')
ax.set_xlabel('연월일')

ax.legend()
plt.show()

```

# 3. 그래프로 자료표현하기(단위가 다른경우)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

#데이터 불러오기
f=open('대기정보측정자료_180825.csv')
data=pd.read_csv(f,index_col=0, parse_dates=['날짜'])

#ppm단위의 그래프
fig=plt.figure(figsize=(10,4))
ax=fig.add_subplot(111)
ax.plot(data.index, data['오존'], label='오존',color='red')
ax.plot(data.index, data['일산화탄소'], label='일산화탄소',color='blue')
ax.plot(data.index, data['이산화질소'], label='이산화질소',color='green')
ax.plot(data.index, data['아황산가스'], label='아황산가스',color='yellow')
#micro단위의 그래프
ax2=ax.twinx()
ax2.plot(data.index, data['PM10'], label='PM10',color='pink')
ax2.plot(data.index, data['PM2.5'], label='PM2.5',color='gold')

#축의 정보설정
ax.set_title('대기정보측정자료')
ax.set_ylabel('ppm')
ax.set_xlabel('연월일')
ax2.set_ylabel('microgram')

#범례의 유효화
ax.legend(loc=2)
ax2.legend(loc=1)
plt.show()

```

# 4. 산포도로 자료표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')

#데이터 불러오기
f=open('대기정보측정자료_180825.csv')
data=pd.read_csv(f,index_col=0, parse_dates=['날짜'])

fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(data.index, data['오존'], label='오존', color='red')

ax2=ax.twinx()
ax2.plot(data.index, data['PM10'], label='PM10',color='blue')

ax.set_title('대기정보측정자료')
ax.set_ylabel('ppm')
ax.set_xlabel('연월일')
ax2.set_ylabel('microgram')

#범례의 유효화
ax.legend(loc=2)
ax2.legend(loc=1)
plt.show()

```
