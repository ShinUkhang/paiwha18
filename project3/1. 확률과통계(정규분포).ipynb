1. 히스토그램으로 표현하기

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

#데이터 조건
mu=100
sigma=10
np.random.seed(0)
x=np.random.normal(mu,sigma,10000)

#히스토그램으로 표현하기
fig=plt.figure()
ax=fig.add_subplot(111)
ax.hist(x)

plt.show()
```

4. 히스토그램과 확률밀도함수를 동시에 보기

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')

#데이터 조건
mu=100
sigma=10
np.random.seed(0)
x=np.random.normal(mu,sigma,10000)

#히스토그램으로 표현하기
fig=plt.figure()
bins=100
ax1=fig.add_subplot(111)
ax1.hist(x,rwidth=0.7, bins=bins,color='blue',label='히스토그램')
ax1.set_title('히스토그램과 정규분포') #제목 넣기
ax1.set_ylabel('hist') #y축 표시

#정규분포그래프 그리기
ax2=ax1.twinx()
x=np.linspace(60,140)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x2-mu)**2 / (2*sigma**2))
ax2.plot(x,y,color='red',label='정규분포')
ax2.set_ylabel('normal') #y축표시

#범례설정
ax1.legend(loc=1)
ax2.legend(loc=2)

plt.show()
```
