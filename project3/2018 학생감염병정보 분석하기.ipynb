```python
import csv
import matplotlib.pyplot as plt

f=open('2018학생감염병정보.csv')
data=csv.reader(f)

r1=[]
r2=[]
r3=[]
r4=[]
r5=[]
r6=[]
for row in data:
    if '구분' not in row:
        r1.append(row[0].replace('주',''))
        r2.append(float(row[1]))
        r3.append(float(row[2]))
        r4.append(float(row[3]))
        r5.append(float(row[4]))
        r6.append(float(row[5]))

plt.figure(figsize=(10,5))
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.plot(r1,r2, 'r--o', label='유행성이하선염')
plt.plot(r1,r3, 'b--o', label='수두')
plt.plot(r1,r4, 'g--o', label='인플루엔자')
plt.plot(r1,r5, 'm--o', label='수족구병')
plt.plot(r1,r6, 'y--o', label='유행성각결막염')
plt.xlabel('주-시간')
plt.ylabel('10만명당 인원')
plt.legend()
plt.title('2018 학생감염병정보-학생건강정보센터')
plt.savefig('감염병')
plt.show()
```
