```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rc('font',family='Malgun Gothic')

f=open('학생몸무게자료.csv')
data=csv.reader(f)
result=[]

for row in data:
    if '남자' in row:
        for i in range (2, len(row)):
            result.append(float(row[i]))

mu=np.mean(result)
sigma=np.std(result)

x=np.linspace(170,177)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--')
plt.xlim(170,177)
plt.title('정규분포')
plt.savefig('첨부사진2')
plt.show()


```

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rc('font',family='Malgun Gothic')

f=open('의료기관분포.csv')
data=csv.reader(f)
result=[]

for row in data:
    if '지역별(1)' not in row:
        result.append(float(row[3])/float(row[2]))
mu=np.mean(result)
sigma=np.std(result)

plt.hist(result, bins=40, density=True, edgecolor='k', linewidth=1.2, alpha=0.4)

x=np.linspace(0.1,0.5)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--', color='g')
plt.xlim(0.1,0.5)
plt.title('전국병원분포비율')
plt.show()
print(mu,sigma)
```

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
plt.rc('font',family='Malgun Gothic')
plt.figure(dpi=100)

f=open('seoul_180803.csv')
data=csv.reader(f)
next(data) #두번째 줄부터 데이터로 인식

result=[]
#11월 달의 기온을 110년간 추출
for row in data:
    if row[-3] != '' and row[0].split('-')[1] == '11' :
        result.append(float(row[-3]))
#평균과 분산
mu=np.mean(result)
sigma=np.std(result)
#히스토그램
plt.hist(result, bins=40, density=True, edgecolor='k', linewidth=1.2, alpha=0.4)

#정규분포
x=np.linspace(-10,20)
y=1/(sigma * np.sqrt(2 * np.pi))*np.exp(-(x-mu)**2 / (2*sigma**2))
plt.plot(x,y, color='red')
plt.axvline(x=mu, linestyle='--', color='g')
plt.title('11월 기온과 정규분포')
plt.savefig('11월 기온분포')
plt.show()
print(mu,sigma)
```
