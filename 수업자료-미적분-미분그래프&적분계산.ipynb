```python
import numpy as np 
import matplotlib.pyplot as plt
from sympy import * #미적분과 같은 수학계산을 할 때 필요합니다.
plt.rc('font',family='HYGothic-Extra',size=12) #글자체와 크기를 설정합니다. 
plt.rcParams['axes.unicode_minus']=False #음수표현이 가능합니다.
#함수
x=np.linspace(-1,3,1000) #-1부터 3까지 1000개의 구간으로 나누어줍니다.
f=lambda x:x*sin(x**2)+1 #람다함수는 수학식을 표현할 때 활용, f(x)=x*sin(x**2)+1로 대체한다.
y=[f(i) for i in x]
#도함수
t=symbols('t') #수식작업 때 활용할 변수를 이렇게 표현하기도 한다. 
der = diff(t*sin(t**2)+1,t)
y2=[der.evalf(subs={t:i}) for i in x] #t 대신에 x를 대입해서 미분한다.
#이계도함수
s=symbols('s') #수식작업 때 활용할 변수를 이렇게 표현하기도 한다. 
der = diff(s*sin(s**2)+1,s,2)
y3=[der.evalf(subs={s:i}) for i in x] #s 대신에 x를 대입해서 미분한다.


#그래프의 작성
plt.plot(x,y, color='b', label='f(x)')
plt.plot(x,y2,color='r',label='f(x)미분')
plt.plot(x,y3,color='g', label='f(x) 이계도함수')

#x축과 y축의 설정
plt.axvline(x=0, color='k')
plt.axhline(y=0,color='k')
plt.style.use('ggplot')
plt.legend(loc='best')
plt.title('f(x)의 그래프와 미분')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from sympy import * #미적분과 같은 수학계산을 할 때 필요합니다.
plt.rc('font',family='HYGothic-Extra',size=12) #글자체와 크기를 설정합니다. 
plt.rcParams['axes.unicode_minus']=False #음수표현이 가능합니다.

x=np.linspace(-1,4,50) #sin그래프를 그려보고 구부분구적법으로 표현하기

f=lambda x:sin(x)/exp(-x)
y=[f(i) for i in x]

#적분값의 계산과 표현
t=symbols('t')
plt.title('0부터 2pi까지 sin(x)/exp(-x) 적분')
print((integrate(sin(t)/exp(-t)),(t,0,2)))

plt.plot(x,y, linewidth=3)
plt.bar(x,y,color='b', alpha=0.9, width=0.1, edgecolor='k')

plt.axvline(x=0, color='k')
plt.axhline(y=0,color='k')
plt.style.use('ggplot')
plt.show()
```
