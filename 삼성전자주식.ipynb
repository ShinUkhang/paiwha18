## 월별로 자료를 리스트에 저장하고 박스로 표현합니다.

import matplotlib.pyplot as plt
import csv

f = open('삼성전자주가.csv')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]] #12개

for row in data :
    if row[-3] != '' and row[-4] != '' :
        for i in range(2,4) :
            month[int(row[0].split('-')[1])-1].append(float(row[i].replace(',','')))
    
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(dpi=300)
plt.boxplot(month, showfliers=False)
plt.show()

## 1월의 자료를 꺽은선 그래프로 표현합니다.

```python
import matplotlib.pyplot as plt
import csv

f = open('삼성전자주가.csv')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]] #12개

for row in data :
    if row[-3] != '' and row[-4] != '' :
        for i in range(2,4) :
            month[int(row[0].split('-')[1])-1].append(float(row[i].replace(',','')))
    
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(dpi=300)
plt.plot(month[0])
plt.show()
```

## 1,2,3월의 자료를 비교하여 그려봅니다. 

```python
import matplotlib.pyplot as plt
import csv

f = open('삼성전자주가.csv')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]] #12개

for row in data :
    if row[-3] != '' and row[-4] != '' :
        for i in range(2,4) :
            month[int(row[0].split('-')[1])-1].append(float(row[i].replace(',','')))
    
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(dpi=300)
plt.plot(month[0],label='1월')#1월 주가등락
plt.plot(month[1],label='2월')#2월 주가등락
plt.plot(month[2],label='3월')#3월 주가등락
plt.legend()
plt.show()
```
